**K-Means Clustering**

```python
X = data.iloc[:,:11].values
```

```python
X
```

```python
from sklearn.cluster import KMeans
```

```python
# finding wcss value for different number of clusters

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(data)
    
    wcss.append(kmeans.inertia_)
```

```python
# plot an elbow graph

sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=0)

# return a label for each data point based on their cluster
Y = kmeans.fit_predict(data)

print(Y)
```

```python
plt.figure(figsize=(15,8))

plt.scatter(X[Y == 0, 0], X[Y == 0, 1], s = 100, c = 'green', label = 'Cluster 1')
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], s = 100, c = 'red', label = 'Cluster 2')
plt.scatter(X[Y == 2, 0], X[Y == 2, 1], s = 100, c = 'yellow', label = 'Cluster 3')
plt.scatter(X[Y == 3, 0], X[Y == 3, 1], s = 100, c = 'violet', label = 'Cluster 4')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'cyan', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income')
plt.ylabel('Age')
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = pd.DataFrame(scaler.fit_transform(data), columns=data.columns)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=11)

principalComponents = pca.fit_transform(data)
```

```python
principalComponents
```

```python
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2',
                         'principal component 3','principal component 4',
                         'principal component 5','principal component 6',
                         'principal component 7','principal component 8',
                         'principal component 9','principal component 10',
                         'principal component 10'])
```

```python
principalDf
```

```python
print(pca.explained_variance_ratio_)
```

```python
loadings = pd.DataFrame(
    pca.components_.T,  # transpose the matrix of loadings
    columns=['principal component 1', 'principal component 2',
                         'principal component 3','principal component 4',
                         'principal component 5','principal component 6',
                         'principal component 7','principal component 8',
                         'principal component 9','principal component 10',
                         'principal component 10'],  # so the columns are the principal components
    index=data.columns,  # and the rows are the original features
)
loadings
```

```python
PC1 = pca.fit_transform(data)[:,0]
PC2 = pca.fit_transform(data)[:,1]

ldngs = pca.components_
```

```python
scalePC1 = 1.0/(PC1.max() - PC1.min())
scalePC2 = 1.0/(PC2.max() - PC2.min())

features = data.columns
```

```python
fig, ax = plt.subplots(figsize=(20, 15))
 
for i, feature in enumerate(features):
    ax.arrow(0, 0, ldngs[0, i], 
             ldngs[1, i])
    ax.text(ldngs[0, i] * 1.15, 
            ldngs[1, i] * 1.15, 
            feature, fontsize=18)
 
ax.scatter(PC1 * scalePC1,PC2 * scalePC2)
 
ax.set_xlabel('PC1', fontsize=20)
ax.set_ylabel('PC2', fontsize=20)
ax.set_title('Figure 1', fontsize=20)
```
